```python
demo_df   = pd.read_csv("./LPOINT_BIG_COMP_01_DEMO.csv")      # 고객 데모정보
pdde_df   = pd.read_csv("./LPOINT_BIG_COMP_02_PDDE.csv")      # 상품 구매 정보 
cop_df    = pd.read_csv("./LPOINT_BIG_COMP_03_COP_U.csv")     # 제휴사 이용 정보
pd_df     = pd.read_csv("./LPOINT_BIG_COMP_04_PD_CLAC.csv")   # 상품 분류 정보 
br_df     = pd.read_csv("./LPOINT_BIG_COMP_05_BR.csv")        # 점포 정보
lpay_df   = pd.read_csv("./LPOINT_BIG_COMP_06_LPAY.csv")      # 엘페이 이용
```

## 온 오프라인으로 나눠서 분석 진행

- 각 제휴사별 판매 물품
- 각 제휴사별 주요 연령

# 전체 제휴사 EDA

```python
# 제휴사별 온 오프라인 데이터 프레임 분리
a_off = {}
a_on = {}
for i in range(1,7):
    t = pd.merge(demo_df, pdde_df[pdde_df['cop_c'] == f'A0{i}'], how='inner', on='cust')
    
    # 임시 제휴사의 오프라인 데이터
    b_off = t[t['chnl_dv']==1].reset_index(drop=True)
    # 임시 제휴사의 온라인 데이터
    b_on = t[t['chnl_dv']==2].reset_index(drop=True)
    
    c_off = pd.merge(b_off, pd_df, how='inner', on='pd_c')
    c_on = pd.merge(b_on, pd_df, how='inner', on='pd_c')

    a_off[f"a{i}"] = c_off.reset_index(drop=True)
    a_on[f"a{i}"]  = c_on.reset_index(drop=True)
```

# 유통사별 구매 건수 / 구매액 확인 (오프라인 데이터 확인)

```python
fig, ax = plt.subplots(6, 1, figsize=(18,12))
for i in range(6):
    try:
        an_plot_data = a_off[f"a{i+1}"].groupby('clac_hlv_nm').sum()['buy_ct'].reset_index().sort_values(by='buy_ct',ascending=False).head(10)
        
        sns.barplot(x='clac_hlv_nm', y="buy_ct", data=an_plot_data, ax=ax[i])
        ax[i].set(xlabel=None, fontsize=15)
        ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=90, fontsize=15)
        ax[i].set_title(f"a{i+1} 제휴사 품목별 구매 건수")
    except:
        ax[i].set(xlabel=None)
        ax[i].set_title(f"a{i+1} 제휴사 품목별 구매 건수")
        ax[i].set_xticklabels(ax[i].get_xticklabels(), fontsize=15)
        
        continue
    
plt.suptitle("유통사별 오프라인 구매 건수", fontsize=20)
plt.tight_layout()
plt.savefig("유통사별 오프라인 구매 건수", facecolor='#FFFFFF')
plt.show()
```
