```python
demo_df = pd.read_csv("LPOINT_BIG_COMP_01_DEMO.csv")
pdde_df = pd.read_csv("LPOINT_BIG_COMP_02_PDDE.csv")
cop_df = pd.read_csv("LPOINT_BIG_COMP_03_COP_U.csv")
pd_df = pd.read_csv("LPOINT_BIG_COMP_04_PD_CLAC.csv")
br_df = pd.read_csv("LPOINT_BIG_COMP_05_BR.csv")
lpay_df = pd.read_csv("LPOINT_BIG_COMP_06_LPAY.csv")
```

```python
cus_merge.drop(index2, axis=0, inplace=True)
cus_merge.drop(index3, axis=0, inplace=True)
cus_merge.drop(index4, axis=0, inplace=True)
cus_merge.drop(index5, axis=0, inplace=True)
cus_merge.drop(index6, axis=0, inplace=True)
cus_merge.drop(index7, axis=0, inplace=True)
cus_merge.drop(index8, axis=0, inplace=True)
```

```python
# 타켓으로 생각한 304050대 분들
cus_304050 = cus_merge.loc[((cus_merge['ages'] == '30대') | (cus_merge['ages'] == '40대')| (cus_merge['ages'] == '50대')) & ((cus_merge['cop_c'] == 'A01') |  (cus_merge['cop_c'] == 'A06')) , : ]
cus_304050
```

```python
from datetime import datetime, date
cus_304050['de_dt'] = pd.to_datetime(cus_304050['de_dt'],format='%Y%m%d')
```

## 직장동료

```python
# 주말과 공휴일 제외하고 12시 이전과 오후 1시부터 6시 사이를 근무시간이라 생각해 그 때 오프라인 구매하신 분들을 제외
q = cus_304050[cus_304050['chnl_dv'] == 1]
qq = q[((q['de_hr'] >= 12) & (q['de_hr'] <= 13)) | (q['de_hr'] >= 18)]
qqq = qq[(qq['de_dt'].dt.dayofweek != 5) | (qq['de_dt'].dt.dayofweek != 6)]
qqqq = qqq[(qqq['de_dt'] == '2021-01-01')|(qqq['de_dt'] == '2021-02-11')|(qqq['de_dt'] == '2021-02-12') \
                   |(qqq['de_dt'] == '2021-03-01')|(qqq['de_dt'] == '2021-05-05')|(qqq['de_dt'] == '2021-05-19')|(qqq['de_dt'] == '2021-09-20')|(qqq['de_dt'] == '2021-09-21')|(qqq['de_dt'] == '2021-09-22')]
qqqq
```

```python
worker = qqq.drop(qqqq.index)
worker
```

```python
# 필터링한 분들 중에서 정장류를 구매하신분들이라면 직장인일 확률이 높을 것이다
display(worker.loc[(worker['pd_nm'] == '남성정장셔츠') | (worker['pd_nm'] == '남성정장화') | (worker['pd_nm'] == '여성정장세트')| (worker['pd_nm'] == '남성정장세트')| (worker['pd_nm'] == '남성정장재킷')| (worker['pd_nm'] == '남성정장바지'),:])
cust_woker = worker.loc[(worker['pd_nm'] == '남성정장셔츠') | (worker['pd_nm'] == '남성정장화') | (worker['pd_nm'] == '여성정장세트')| (worker['pd_nm'] == '남성정장세트')| (worker['pd_nm'] == '남성정장재킷')| (worker['pd_nm'] == '남성정장바지'),:]['cust'].unique()
```

```python
worker_suit = cus_304050[cus_304050['cust'].isin(cust_woker)]
worker_suit
```
